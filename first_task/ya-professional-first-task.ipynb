```python
import re
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from pymystem3 import Mystem
import nltk

lemmatizer = Mystem()
def preprocessing(line, token = lemmatizer):
    line = line.lower()
    line = re.sub(r"[{}]".format(string.punctuation), " ", line)
    line = ' '.join(token.lemmatize(line))
    line = line.replace('\n\n', ' ').replace('\n', ' ')
    return line

nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
def tfidf_vectorizing(df, col_name, index_col):
    russian_stopwords = stopwords.words('russian')
    tfidf = TfidfVectorizer(
        stop_words = russian_stopwords[:50],
        preprocessor = preprocessing
    )
    tfidf_data = tfidf.fit_transform(df[col_name]).toarray()
    tfidf_data = pd.DataFrame(
        tfidf_data,
        index = index_col,
        columns = tfidf.get_feature_names_out()
    )
    return tfidf_data
```

```python
tfidf_data_train = tfidf_vectorizing(X_train, 'Заголовок', X_train.index)
tfidf_data_train
```

```python
def make_new_columns_tfidf(df, tfidf_df):
    df['SumTfidf'] = tfidf_df.sum(axis = 1).reset_index()[0]
    df['MaxTfidf'] = tfidf_df.max(axis = 1).reset_index()[0]
    df['Meanfidf'] = tfidf_df.mean(axis = 1).reset_index()[0]

make_new_columns_tfidf(X_train, tfidf_data_train)
X_train
```

```python
from sklearn.decomposition import PCA

def make_clusters(df, tfidf_data):
    centered = tfidf_data - tfidf_data.mean()
    pca = PCA(n_components = 20)
    
    pca_decomp = pca.fit_transform(centered)
    
    from sklearn.cluster import KMeans
    
    kmeans = KMeans(n_clusters = 15, random_state = 0)
    kmeans.fit(pca_decomp)
    
    dist_cols = ['DistanceTo1thCluster',
     'DistanceTo2thCluster',
     'DistanceTo3thCluster',
     'DistanceTo4thCluster',
     'DistanceTo5thCluster',
     'DistanceTo6thCluster',
     'DistanceTo7thCluster',
     'DistanceTo8thCluster',
     'DistanceTo9thCluster',
     'DistanceTo10thCluster',
     'DistanceTo11thCluster',
     'DistanceTo12thCluster',
     'DistanceTo13thCluster',
     'DistanceTo14thCluster',
     'DistanceTo15thCluster'
    ]
    
    cluster_distances = pd.DataFrame(
        data = kmeans.transform(pca_decomp),
        columns = dist_cols
    )
    return pd.concat((df, cluster_distances), axis = 1)

X_train = make_clusters(X_train, tfidf_data_train)
```

```python
def date_agg(df, col_name):
    df['day'] = pd.to_datetime(df[col_name]).dt.day
    df['month'] = pd.to_datetime(df[col_name]).dt.month
    df['year'] = pd.to_datetime(df[col_name]).dt.year

def feature_changing(df):
    date_agg(df, 'Дата')
    df = df.drop(['№ Объявления', 'Заголовок', 'Текст'], axis = 1)
    df = df.drop('Дата', axis = 1)
    return df

X_train = feature_changing(X_train)
```

```python
from category_encoders import TargetEncoder
from category_encoders.one_hot import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

def columns_division(df):
    ohe_cols = [x for x in df.columns if df[x].nunique()<8 ]
    mte_cols = [x for x in df.columns if df[x].nunique()>=8]
    numeric_cols = list(df.select_dtypes(exclude = 'object').columns)
    
    ohe_cols_idx = [list(df.columns).index(col) for col in ohe_cols]
    mte_cols_idx = [list(df.columns).index(col) for col in mte_cols]
    numeric_cols_idx = [list(df.columns).index(col) for col in numeric_cols]

    return ohe_cols_idx, mte_cols_idx, numeric_cols_idx
```

```python
ohe_cols_idx, mte_cols_idx, numeric_cols_idx = columns_division(X_train)
```

```python
# трансформация
t = [('OneHotEncoder', OneHotEncoder(), ohe_cols_idx),
    ('MeanTargetEncoder', TargetEncoder(), mte_cols_idx),
    ('StandardScaler', StandardScaler(), numeric_cols_idx)]
```

```python
transformer = ColumnTransformer(t)
transformer.fit(X_train, y_train)
transformer.transform(X_train)
```

## Построим логистическую регрессию

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

pipe_log_reg = Pipeline([('column_transformer', transformer),
                ('log_reg', LogisticRegression())])

pipe_log_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
print(f'train score is {accuracy_score(pipe_log_reg.predict(X_train), y_train)}')
```

```python
log_reg_predictions = pipe_log_reg.predict(X_test)
pd.DataFrame(log_reg_predictions).to_csv('new_log_reg_predictions.csv', index = False)
```

## Обучим градиентный бустинг

```python
from xgboost import XGBClassifier
pipe_boosting = Pipeline([('column_transformer', transformer),
                ('boosting', XGBClassifier())])
pipe_boosting.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'boosting__n_estimators': [25, 50, 100],
    'boosting__learning_rate': [0.01, 0.05, 0.1],
    'boosting__max_depth': [2, 3, 4]
}

search_boosting = GridSearchCV(pipe_boosting, param_grid,
                              scoring = 'accuracy',
                              cv = 8,
                              n_jobs = -1)
search_boosting.fit(X_train, y_train)

print(f'best score is {search_boosting.best_score_:.5f}')
print(f'params for best score are {search_boosting.best_params_}')
```

```python
boosting_predictions = search_boosting.predict(X_test)
pd.DataFrame(boosting_predictions).to_csv('new_boosting_model.csv', index = False)
```

## Обучим метрическую модель - KNN

```python
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
pipe_knn = Pipeline([('column_transformer', transformer),
                    ('KNN', KNeighborsClassifier())])
pipe_knn.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
knn_param_grid = {
    'KNN__n_neighbors': [5, 7, 8, 10],
    'KNN__weights': ['uniform', 'distance']
}
search_knn = GridSearchCV(pipe_knn, 
                          knn_param_grid,
                         scoring = 'accuracy',
                         cv = 8,
                         n_jobs = -1)
search_knn.fit(X_train, y_train)
```

```python
print(f'best score is {round(search_knn.best_score_,5)}')
print(f'best params are: {search_knn.best_params_}')
```

```python
knn_predictions = search_knn.predict(X_test)
pd.DataFrame(knn_predictions).to_csv('new_knn_predictions.csv', index = False)
```
