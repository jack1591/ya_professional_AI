## Выполним преобразования над текущими колонками

```python
def date_agg(df, col_name):
    df['day'] = pd.to_datetime(df[col_name]).dt.day
    df['month'] = pd.to_datetime(df[col_name]).dt.month
    df['year'] = pd.to_datetime(df[col_name]).dt.year

def feature_changing(df):
    date_agg(df, 'Дата')
    df = df.drop(['№ Объявления', 'Заголовок', 'Текст'], axis = 1)
    df = df.drop('Дата', axis = 1)
```

```python
feature_changing(X_train)
```

## Теперь сформируем колонки для преобразования (one hot encoding, mean target encoding, scaling)

```python
from category_encoders import TargetEncoder
from category_encoders.one_hot import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

def columns_division(df):
    ohe_cols = [x for x in df.columns if df[x].nunique()<8 ]
    mte_cols = [x for x in df.columns if df[x].nunique()>=8]
    numeric_cols = list(df.select_dtypes(exclude = 'object').columns)
    
    ohe_cols_idx = [list(df.columns).index(col) for col in ohe_cols]
    mte_cols_idx = [list(df.columns).index(col) for col in mte_cols]
    numeric_cols_idx = [list(df.columns).index(col) for col in numeric_cols]

    return ohe_cols_idx, mte_cols_idx, numeric_cols_idx
```

```python
ohe_cols_idx, mte_cols_idx, numeric_cols_idx = columns_division(X_train)
```

```python
# трансформация
t = [('OneHotEncoder', OneHotEncoder(), ohe_cols_idx),
    ('MeanTargetEncoder', TargetEncoder(), mte_cols_idx),
    ('StandardScaler', StandardScaler(), numeric_cols_idx)]
```

```python
transformer = ColumnTransformer(t)
transformer.fit(X_train, y_train)
transformer.transform(X_train)
```

## Построим модель логистической регрессии

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
pipe = Pipeline([('column_transformer', transformer),
                ('log_reg', LogisticRegression())])
pipe.fit(X_train, y_train)
log_reg_predictions = pipe.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
print(f'Значение accuracy для логистической регрессии равно {round(accuracy_score(log_reg_predictions, Y_train),3)}')
```
